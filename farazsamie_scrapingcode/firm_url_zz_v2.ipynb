#### I am trying to scrape the web links for the payday loan firms listed on this website. There are 15 pages and the code should loop through the page and print the URLS listed on each page here.

#### https://www.bbb.org/us/category/payday-loans?page=1

#### The issue is that everytime you reload the page, the urls change. There are 7,000 firms that populate but about 200-ish are listed on the 15 pages and there is no way to know if one firm is going to printed twice. 

### This is the second page attempting to scrape the URLS -- Need to cross reference these from the version1 csv to see if the URLS are different. 


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Set up the Selenium WebDriver (using Chrome as an example)
driver = webdriver.Chrome()

# Function to fetch firm URLs from a given page
def get_firm_urls(page_url, seen_urls):

    """
    Fetches all firm URLs from a given page using Selenium to handle CAPTCHA and dynamic content.
    Avoids duplicates using the provided `seen_urls` set.
    """

    driver.get(page_url)
    print(f"Accessing page: {page_url}")
    
    # Wait for the user to solve CAPTCHA manually, if prompted
    input("If prompted, please solve the CAPTCHA and press Enter to continue...")
    
    # Allow time for the page to fully load
    time.sleep(5)  

    # Parse the rendered page with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Debugging: Print a portion of the loaded page to inspect its structure
    
    # Print the first 1000 characters of the page
    print(soup.prettify()[:1000])  

    # Extract firm URLs from the page
    firm_urls = []
    for link in soup.find_all('a', href=True):

        # Adjust pattern as necessary for the specific structure
        if '/profile/' in link['href']:  
            full_url = 'https://www.bbb.org' + link['href']

    #          #Check if the URL not already seen. Add URLs that haven't been seen
            if full_url not in seen_urls:  

                # List new URL
                print(f"Found new URL: {full_url}")  
                firm_urls.append(full_url)

    #            #add URL to seen_urls list
                seen_urls.add(full_url)  
            else:
    #            #skip duplicates
                print(f"Duplicate URL skipped: {full_url}")  
    return firm_urls

# Main scraping loop
base_url = 'https://www.bbb.org/us/category/payday-loans?page='

    #   #Store all collected firm URLs
all_firm_urls = []  

    #   #track urls that have been processed. 
seen_urls = set()   


for page in range(1, 16):  # Loop through pages 1 to 15
        page_url = base_url + str(page)
        firm_urls = get_firm_urls(page_url, seen_urls)  # Pass the global `seen_urls` set
        all_firm_urls.extend(firm_urls)  # Add the new URLs to the main list
        time.sleep(2)  # Delay between page requests

    # Print the total number of unique firm URLs collected
print(f"Total unique firm URLs collected: {len(all_firm_urls)}")
for url in all_firm_urls:
        print(url)


# Close the WebDriver when done
driver.quit()

```

```python
import csv

# Save all of the printed URLs to a CSV file
csv_filename = 'firm_urls_v2.csv'
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Firm URL'])  # Header row
    for url in all_firm_urls:
        writer.writerow([url])

print(f"Total firm URLs collected: {len(all_firm_urls)}")
print(f"URLs saved to {csv_filename}")
```
