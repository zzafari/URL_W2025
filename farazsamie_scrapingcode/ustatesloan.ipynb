```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time


driver = webdriver.Chrome()

# extract details from TitleMax page
def tm_offices(lender_url):
    # Navigate to the page and let it load
    driver.get(lender_url)
    time.sleep(10)  

    # Wait for user to solve CAPTCHA manually
    input("Solve CAPTCHA if prompted, then press Enter to continue...")

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Find the table 
    table = soup.find('table', class_='u-datatable')
    
    if not table:
        print("Error 1: Could not find the table.")
        return []

    # Find the tbody element 
    tbody = table.find('tbody', class_='font-size-1')
    
    if not tbody:
        print("Error 2: Could not find tbody with class 'font-size-1'.")
        return []

    # Extract each row
    rows = tbody.find_all('tr')
    
    office_details = []

    # Iterate over each row
    for row in rows:
        # Find all td elements in this row
        cols = row.find_all('td')
        
        if len(cols) > 0:
            # Extract office address 
            address_div = cols[0].find('div')
            address = address_div.get_text(strip=True) if address_div else None
            
            # Extract phone number from the 'tel' 
            phone_div = cols[0].find_all('div')[1] if len(cols[0].find_all('div')) > 1 else None

            if phone_div and phone_div.get('tel'): 
                tel = phone_div.get('tel')
            else:
                tel = None
            
            
            if address and tel:
                print(f"Address: {address}, Tel: {tel}")
                office_details.append({
                    'Office Address': address,
                    'Tel': tel
                })

    return office_details


# Lender's URL for TitleMax
lender_url = "https://www.ustatesloans.org/lender/titlemax/#t/"

# Loop 
page_number = 1
all_office_data = []

while True:
    print(f"Scraping page {page_number}...")
    
    # scrape data from the current page
    office_data = tm_offices(lender_url)
    if not office_data:
        break  # Stop if no more data is found

    # Add the office data to the final list
    all_office_data.extend(office_data)

    # Check if there is a "Next" button to go to the next page
    try:
        next_button = driver.find_element(By.ID, 'DataTables_Table_O_next')
        # Click "Next" if the button is enabled
        if next_button.is_enabled():  
            next_button.click()
            time.sleep(10)  
            page_number += 1
        else:
            # Exit if no "Next" button or it's disabled
            break  
    except Exception as e:
        break


driver.quit()
```

```python
import csv

csv_filename = 'ustatesloan_titlemax_office_info.csv'
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Office Address', 'tel'])  # Write the header row

    for office in all_office_data:
        writer.writerow([office['Office Address'], office['tel']])

print(f"Total office information collected: {len(office_data)}")
print(f"Office information saved to {csv_filename}")
driver.quit()
```
