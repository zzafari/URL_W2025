```python
#imports

import pandas as pd
import re
import numpy as np
```

```python
#read in webscraped data

with open('payday_loan_data.txt', 'r', encoding="utf8") as file:
    text = file.read()
```

```python
#gets row of all states
rows = re.findall(r'========================================\n\nState:(?:.|\n)*?History of Payday Loans:', text)
```

```python
df = pd.DataFrame(columns=['State'])
```

```python
for row in rows:
    state = re.findall(r'State: (?:.|\n)*?\n=', row)[0][7:-2]
    paydayData = re.findall(r'Payday Loan Regulations:(?:.|\n)*?\nHistory', row)[0]
    rowData = paydayData.splitlines()

    df.loc[len(df)] = np.nan
    df.loc[len(df)-1, 'State'] = state

    if len(rowData) < 3:
        break

    for tuple in rowData[1:-2]:
        attr = tuple.split(':')[0]
        data = tuple.split(':')[1][1:]
        if attr in df.columns:
            df.loc[len(df)-1, attr] = data
        else:
            df[attr] = np.nan
            df.loc[len(df)-1, attr] = data
```
